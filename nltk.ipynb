```python
from nltk import word_tokenize, sent_tokenize

sentence = """ John walked to the store. He bought some milk and bread. Then, he returned home."""

words = word_tokenize(sentence)
sent = sent_tokenize(sentence)

print(words)
print(sent)
```


Stemming


```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
from nltk import PorterStemmer

pstemmer = PorterStemmer()

for word in words:
    print(word + " --> " + pstemmer.stem(word))
```

```python
pstemmer.stem("communication") #meaningless
```

```python
from nltk.stem import RegexpStemmer

reg = RegexpStemmer("ing$|s$|e$|able$", min=4)

print("cars --> " + reg.stem("cars"))
print("eating --> " +reg.stem("eating"))
print("tracable --> " +reg.stem("tracable"))
print("apple --> " +reg.stem("apple"))
```

```python
from nltk.stem import SnowballStemmer

snow = SnowballStemmer("english")

for word in words:
    print(word + " --> " + snow.stem(word))

```

```python
pstemmer.stem("fairly"),pstemmer.stem("sportingly")
```

```python
snow.stem("fairly"),snow.stem("sportingly"), snow.stem("goes")
```

Wordnet Lemmatizer

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()


for word in words:
    print(word + " --> " + lemma.lemmatize(word,pos='v'))
```

```python

lemma.lemmatize("goes",pos='v') ,lemma.lemmatize("fairly",pos='v'),lemma.lemmatize("sportingly")
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
sents = nltk.sent_tokenize(paragraph)
```

```python
for i in range(len(sents)):
    words = nltk.word_tokenize(sents[i])
    words = [pstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]

    sents[i] = ' '.join(words)

sents
```
